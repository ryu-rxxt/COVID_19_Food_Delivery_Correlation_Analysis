```python
#library import

import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

from matplotlib import pyplot as plt
import seaborn as sns
sns.set()

#from scipy.constants import precision

import statsmodels.formula.api as smf
import statsmodels.api as sm

%precision 3

%matplotlib inline
```

```python
#read data

food_data = pd.read_csv("온라인음식서비스거래액.csv", engine="python", thousands = ',')
covid_data = pd.read_csv("코로나확진현황.csv", engine="python", thousands = ',')
covid_m_data = pd.read_csv("코로나확진현황(월간).csv", engine="python", thousands = ',')

food_data['date'] = pd.to_datetime(food_data['date'], format='%Y/%m')
covid_data['date'] = pd.to_datetime(covid_data['date'])
covid_m_data['date'] = pd.to_datetime(covid_m_data['date'], format='%Y.%m')
```

```python
# 일별 확진자 graph(1) (2020.1.20~2021.2.22)

x = covid_data["date"][:100]
y = covid_data["infected"][:100]
plt.plot(x,y,color='blue')

x = covid_data["date"][100:200]
y = covid_data["infected"][100:200]
plt.plot(x,y,color='orange')

x = covid_data["date"][200:300]
y = covid_data["infected"][200:300]
plt.plot(x,y,color='blue')

x = covid_data["date"][300:400]
y = covid_data["infected"][300:400]
plt.plot(x,y,color='orange')

plt.title("Infection graph over time\n(2020.1.20~2021.2.22)")
plt.xlabel("date")
plt.ylabel("infected")
```

```python
# 월별 신규확진자 graph

x = covid_m_data["date"]
y = covid_m_data["infected"]

plt.plot(x,y,color='blue')
plt.title("new infection per month\n(2020.01~2022.05)")

plt.xlabel("date")
plt.ylabel("infected")
```

```python
# 월별 온라인 음식서비스 매출 graph / bar

x = food_data["date"]

y = food_data["total"]
plt.plot(x, y, color = "red")

y = food_data["general"]
plt.plot(x, y, color = "green")

y = food_data["specific"]
plt.plot(x, y, color = "blue")

plt.plot(food_data, kind='bar')
```

```python
# 월별 신규확진자-배달매출액 scatter plot

merge = pd.merge(covid_m_data, food_data, on = 'date')
print(merge)


sns.relplot(x='infected', y='total', data = merge)
```
